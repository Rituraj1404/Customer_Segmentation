```python
#Handling Outliers

num_cols = df.select_dtypes(include=['float64', 'int64']).columns


for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

print(df.head())
```

# K-Means Clustering

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters= i, init = 'k-means++',random_state= 42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS value')
plt.show()
```

```python
kmeans = KMeans(n_clusters= 3, init = 'k-means++',random_state= 42)
y_kmeans = kmeans.fit_predict(X)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import numpy as np

pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
y_kmeans = kmeans.fit_predict(X_reduced)

colors = ['red', 'blue', 'green', 'cyan']

plt.figure(figsize=(10, 6))
for i in range(n_clusters):
    plt.scatter(X_reduced[y_kmeans == i, 0], X_reduced[y_kmeans == i, 1], s=100, c=colors[i], label=f"Cluster {i + 1}")

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', marker='X', label='Centroids')

plt.title('K-means Clustering (PCA-reduced Data)')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.grid(True)

plt.show()
```

```python
components = pca.components_

# Get the explained variance ratio (how much variance each principal component explains)
explained_variance = pca.explained_variance_ratio_

# Print the explained variance for each component
print("Explained variance by each component:", explained_variance)

# Find the top contributing features for each principal component
for i in range(pca.n_components_):
    feature_contributions = components[i]
    feature_contribution_df = pd.DataFrame({
        'Feature': X.columns,
        'Contribution': feature_contributions
    }).sort_values(by='Contribution', ascending=False)

    print(f"\nTop contributing features for component {i+1} (variance explained: {explained_variance[i]:.2f}):")
    print(feature_contribution_df)

# To get the two most important features across both components
# Sum the absolute contributions of each feature across all components
absolute_contributions = np.abs(components).sum(axis=0)

# Sort the features by their total contribution
important_features = pd.DataFrame({
    'Feature': X.columns,
    'Total Contribution': absolute_contributions
}).sort_values(by='Total Contribution', ascending=False)

print("\nTop 2 most important features across both components:")
print(important_features.head(2))
```

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(X_reduced, y_kmeans)
print(f"Silhouette Score: {sil_score}")
```

# Heirarichal Clustering

```python
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(X,method = 'ward'))
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from scipy.cluster.hierarchy import fcluster

linked = sch.linkage(X_reduced, method='ward')

y_hierarchical = fcluster(linked, t=2, criterion='maxclust')

colors = ['red', 'blue']

plt.figure(figsize=(10, 6))
for i in range(2):
    plt.scatter(X_reduced[y_hierarchical == i + 1, 0], X_reduced[y_hierarchical == i + 1, 1], s=100, c=colors[i], label=f"Cluster {i + 1}")

for i in range(2):
    centroid = X_reduced[y_hierarchical == i + 1].mean(axis=0)
    plt.scatter(centroid[0], centroid[1], s=300, c='yellow', marker='X', label=f'Centroid {i + 1}')

plt.title('Hierarchical Clustering (PCA-reduced Data)')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.grid(True)

plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(X_reduced, y_hierarchical)
print(f"Silhouette Score: {sil_score}")
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import numpy as np

dbscan = DBSCAN(eps=1, min_samples=5)
y_dbscan = dbscan.fit_predict(X_reduced)

if len(np.unique(y_dbscan)) > 1 and -1 in y_dbscan:
    valid_labels = y_dbscan != -1
    score = silhouette_score(X_reduced[valid_labels], y_dbscan[valid_labels])
elif len(np.unique(y_dbscan)) > 1:
    score = silhouette_score(X_reduced, y_dbscan)
else:
    score = -1

if score != -1:
    print(f"Silhouette Score: {score}")
else:
    print("Silhouette Score is undefined (less than 2 clusters or only noise points).")
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import numpy as np

eps_values = np.arange(0.5, 3.0, 0.5)
min_samples_values = range(3, 10)

best_score = -1
best_eps = None
best_min_samples = None

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        y_dbscan = dbscan.fit_predict(X_reduced)

        unique_labels = set(y_dbscan)
        if len(unique_labels - {-1}) > 1:
            valid_labels = y_dbscan != -1
            score = silhouette_score(X_reduced[valid_labels], y_dbscan[valid_labels])

            if score > best_score:
                best_score = score
                best_eps = eps
                best_min_samples = min_samples

if best_score != -1:
    print(f"Best Silhouette Score: {best_score}")
    print(f"Optimal Parameters - eps: {best_eps}, min_samples: {best_min_samples}")
else:
    print("No valid clusters found with the given parameter ranges.")
```
